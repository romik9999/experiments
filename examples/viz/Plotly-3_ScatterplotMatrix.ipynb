# go.Splom
https://plot.ly/python/splom/

```python
TARGET_COL = 'churn'

classes=np.unique(df[TARGET_COL].values).tolist()
class_code={classes[k]: k for k in range(len(classes))}
color_vals=[class_code[cl] for cl in df[TARGET_COL]]

pl_colorscale=[[0.0, '#19d3f3'],
               [0.333, '#19d3f3'],
               [0.333, '#e763fa'],
               [0.666, '#e763fa'],
               [0.666, '#636efa'],
               [1, '#636efa']]

text=list(df[TARGET_COL].values)
columns = ['total day minutes', 'total night minutes', 
           'total day calls', 'total night calls',
           'customer service calls']

dimensions = [dict(label=col, values=df[col]) for col in columns]

trace1 = go.Splom(dimensions=dimensions,
                  text=text,
                  marker=dict(color=color_vals,
                            size=6,
                            colorscale=pl_colorscale,
                            showscale=False,
                            line=dict(width=0.2,
                                      color='rgb(230,230,230)'))
)
# Comment line below to plot diagonal results
trace1['diagonal'].update(visible=False)
# Comment line below to plot upper half results
trace1['showupperhalf']=False

iplot(dict(data=[trace1]))
```
